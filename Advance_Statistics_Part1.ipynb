1. What is a random variable in probability theory?
  - A random variable in probability theory is a variable whose possible values are numerical outcomes of a random process.

2. What are the types of random variables?
  - Discrete Random Variables: These take on a finite or countable number of distinct values.
  - Continuous Random Variables: These can take on an infinite number of values within a given range.

3. What is the difference between discrete and continuous distributions?
  - Discrete Distributions: Deal with countable outcomes (finite or countably infinite values). Used when data consists of separate, distinct values (e.g., number of defective items in a batch).
  - Continuous Distributions: Deal with uncountable outcomes (infinite values within a range)

4. What are probability distribution functions (PDF)?
  - A Probability Distribution Function (PDF) describes the likelihood of different outcomes for a random variable.

5.  How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
  - Probability Density Function (PDF): Represents the probability of a random variable taking a specific value or falling within a small range.
  - Cumulative Distribution Function (CDF): Represents the probability that a random variable will be less than or equal to a given value.

6. What is a discrete uniform distribution?

8. What is the binomial distribution, and how is it used in probability?
  - The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent Bernoulli trials, each with the same probability of success.

9. What is the Poisson distribution and where is it applied?
  - The Poisson distribution is a discrete probability distribution used to model the number of events occurring in a fixed interval of time or space, assuming that these events happen independently and at a constant average rate.
  - Applications:
    - Queueing systems: Predicting customer arrivals at a store or calls to a call center.
    - Biology: Estimating the number of mutations in DNA strands.

10. What is a continuous uniform distribution?
  - A continuous uniform distribution is a probability distribution where every value within a specified range is equally likely to occur. It is often used when there is no preference or bias toward any particular outcome.

11. What are the characteristics of a normal distribution?
  - A normal distribution, also known as the Gaussian distribution, is one of the most important probability distributions in statistics and data analysis.

12. What is the standard normal distribution, and why is it important?
  - The standard normal distribution is a special case of the normal distribution where the mean (ùúá) is 0, and the standard deviation (ùúé) is 1.
  - Why is it Important?
    - Simplifies Probability Calculations: Since tables exist for the standard normal distribution, probabilities for any normal distribution can be quickly found using Z-scores.
    - Comparison of Different Normal Distributions: Converts different normal distributions into a common scale.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  - The Central Limit Theorem (CLT) is one of the most fundamental concepts in statistics.
  - Why is CLT Critical in Statistics?
    - Basis for Many Statistical Methods: CLT enables hypothesis testing, confidence intervals, and other inferential techniques.
    - Justifies Sampling: Explains why data from a sample can be used to infer conclusions about an entire population.

14. How does the Central Limit Theorem relate to the normal distribution?
  - Relation to the Normal Distribution:
  - Sample Means Follow a Normal Distribution: Even if the original population distribution is not normal, the distribution of sample means will become approximately normal as the sample size increases.
  - Foundation for Normal Approximation: Many real-world data sets follow a normal distribution because they result from the aggregation of many independent random influences‚Äîjust like CLT predicts.

15. What is the application of Z statistics in hypothesis testing?
  - Z-statistics play a crucial role in hypothesis testing, particularly when dealing with population means and proportions.

16.  How do you calculate a Z-score, and what does it represent?
  - A Z-score, also known as a standard score, tells you how far a particular data point is from the mean of a dataset, measured in terms of standard deviations.

17. What are point estimates and interval estimates in statistics?
  - Point Estimate: A single value used to approximate an unknown population parameter.
  - Interval Estimate: Provides a range within which the true population parameter is likely to fall, often with an associated confidence level.

18. What is the significance of confidence intervals in statistical analysis?
  - Confidence intervals are incredibly important in statistical analysis because they provide a measure of uncertainty and reliability in estimating population parameters.

19. What is the relationship between a Z-score and a confidence interval?
  - Z-Scores Quantify Position in a Distribution: A Z-score represents how far a particular data point is from the mean, measured in standard deviations.
  - Confidence Intervals Use Z-Scores to Define Ranges: Confidence intervals estimate where a population parameter (like the mean) is likely to fall.

20. How are Z-scores used to compare different distributions?
  - Z-scores are incredibly useful for comparing different distributions, even when they have different means and standard deviations.

21. What are the assumptions for applying the Central Limit Theorem?
  - The Central Limit Theorem (CLT) is a fundamental principle in statistics, stating that the distribution of the sample mean approaches a normal distribution as the sample size increases‚Äîregardless of the original population‚Äôs shape.
  - Key Assumptions for the Central Limit Theorem
     - Independence of Samples
     - Random Sampling

22. What is the concept of expected value in a probability distribution?
  - Expected value (EV) is a fundamental concept in probability that represents the long-term average outcome of a random variable.

23.  How does a probability distribution relate to the expected outcome of a random variable?
  - A probability distribution defines how likely different outcomes of a random variable are, and the expected value (or expectation) gives the weighted average of those possible outcomes.

```python
#6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
import numpy as np
import matplotlib.pyplot as plt

# Parameters
low = 1  # Minimum value
high = 10  # Maximum value
size = 1000  # Number of samples

# Generate discrete uniform distribution samples
data = np.random.randint(low, high + 1, size)

# Calculate CDF
values, counts = np.unique(data, return_counts=True)
cumulative_counts = np.cumsum(counts) / size  # Normalize for probability

# Plot CDF
plt.step(values, cumulative_counts, where="post", linestyle="--", marker="o", color="blue", label="CDF")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.title(f"CDF of Discrete Uniform Distribution ({low}-{high})")
plt.legend()
plt.grid()

# Show plot
plt.show()
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """Calculate Z-scores for a dataset."""
    mean = np.mean(data)
    std_dev = np.std(data)
    return (data - mean) / std_dev

# Sample dataset
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Generating a normal distribution

# Compute Z-scores
z_scores = calculate_z_scores(data)

# Plot histogram of Z-scores
plt.hist(z_scores, bins=30, density=True, edgecolor='black', alpha=0.7)

# Labels and title
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.title("Z-score Distribution")

# Show plot
plt.show()
```

```python
#16.Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_standard_normal():
    """Plots the standard normal distribution (mean=0, std=1)."""
    x = np.linspace(-4, 4, 1000)  # Generate values from -4 to 4
    y = stats.norm.pdf(x, loc=0, scale=1)  # Calculate probability density function

    # Plot the curve
    plt.plot(x, y, color="blue", label="Standard Normal Distribution")
    plt.xlabel("Value")
    plt.ylabel("Probability Density")
    plt.title("Standard Normal Distribution (mean=0, std=1)")
    plt.legend()
    plt.grid()

    # Show plot
    plt.show()

# Call function to plot
plot_standard_normal()
```

```python
#21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
import numpy as np
import scipy.stats as stats

# Generate synthetic data from a normal distribution
np.random.seed(42)  # For reproducibility
mean_population = 50
std_dev_population = 5
sample_size = 30

sample_data = np.random.normal(mean_population, std_dev_population, sample_size)

# Calculate sample statistics
sample_mean = np.mean(sample_data)
sample_std_dev = np.std(sample_data, ddof=1)  # Using Bessel's correction

# Confidence level
confidence_level = 0.95  # 95% confidence interval
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha/2, df=sample_size - 1)  # Two-tailed t-score

# Margin of error
margin_error = t_critical * (sample_std_dev / np.sqrt(sample_size))

# Confidence interval
lower_bound = sample_mean - margin_error
upper_bound = sample_mean + margin_error

# Output results
print(f"Generated Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# Interpretation
print("\nInterpretation:")
print(f"We are {confidence_level * 100}% confident that the true population mean lies between {lower_bound:.2f} and {upper_bound:.2f}.")
```

```python
#23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Define parameters for the Poisson distribution
lambda_poisson = 5  # Average rate (mean)

# Generate x values (possible number of occurrences)
x_values = np.arange(0, 20)  # Values from 0 to 19

# Compute the CDF
cdf_values = stats.poisson.cdf(x_values, lambda_poisson)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.plot(x_values, cdf_values, marker='o', linestyle='-', color='red', label=f"Poisson CDF (Œª={lambda_poisson})")
plt.xlabel("Number of occurrences")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF) of Poisson Distribution")
plt.legend()
plt.grid()

# Show the plot
plt.show()

# Interpretation: Probability of observing a value <= 5
probability = stats.poisson.cdf(5, lambda_poisson)
print(f"Probability of observing 5 or fewer occurrences: {probability:.4f}")
```

```python
#27. Implement Z-score normalization on a dataset and visualize its transformation.
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate a sample dataset
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, Std Dev=15

# Z-score normalization
mean_data = np.mean(data)
std_dev_data = np.std(data, ddof=1)
z_normalized_data = (data - mean_data) / std_dev_data

# Visualization
plt.figure(figsize=(10, 5))

# Original dataset histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, alpha=0.7, color='blue', label="Original Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Original Dataset")
plt.legend()
plt.grid()

# Z-score normalized dataset histogram
plt.subplot(1, 2, 2)
plt.hist(z_normalized_data, bins=30, alpha=0.7, color='red', label="Z-score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Z-score Normalized Dataset")
plt.legend()
plt.grid()

# Show plots
plt.tight_layout()
plt.show()

# Print sample statistics
print(f"Original Mean: {mean_data:.2f}, Original Std Dev: {std_dev_data:.2f}")
print(f"Normalized Mean: {np.mean(z_normalized_data):.2f}, Normalized Std Dev: {np.std(z_normalized_data, ddof=1):.2f}")
```

```python
#28.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
import numpy as np
import scipy.stats as stats

def calculate_skewness_kurtosis(data):
    """Calculate the skewness and kurtosis of a dataset."""
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)  # Excess kurtosis (subtracts 3)
    return skewness, kurtosis

# Generate synthetic data from a normal distribution
np.random.seed(42)  # For reproducibility
normal_data = np.random.normal(loc=50, scale=10, size=1000)

# Compute skewness and kurtosis
skewness_value, kurtosis_value = calculate_skewness_kurtosis(normal_data)

# Output results
print(f"Skewness: {skewness_value:.4f}")
print(f"Kurtosis (Excess): {kurtosis_value:.4f}")
```
